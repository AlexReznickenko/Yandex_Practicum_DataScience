### Разделяем выборку на обучающую и тестовую 
Так как в дальнейшем будет использоваться GridSearchCV, для подбора лучшей модели понадобится только обучающая выборка, поскольку GridSearchCV использует кросс-валидацию автоматически

```python
features = df.drop('Exited', axis=1)
target = df['Exited']
```

### Применим one hot encoding, избегая ловушки фиктивных призанков

```python
features = pd.get_dummies(features, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split( features, target, test_size=0.25, random_state=42, stratify = target)
```

### Нормализуем данные

```python
scaler=StandardScaler()
scaler.fit(X_train[numeric])
X_train[numeric]=scaler.transform(X_train[numeric])
X_test[numeric]=scaler.transform(X_test[numeric])
```

### Обучаем модели без учёта дисбаланса

```python
#Определяем переменную, которая будет хранить результаты разных моделей для 
#дисбалансированной выборки, чтобы потом сравнить результаты
no_balance_results = []
```

#### Логистическая регрессия

```python
param_grid = {"C": np.arange(0.0, 1.0, 0.1), "penalty":["l2"], 'random_state':[42]}
LR = GridSearchCV( LogisticRegression(random_state = 42), param_grid, 
    cv = 3, scoring = 'f1', verbose = 2)

LR.fit(X_train, y_train)

print("Лучшие гиперпараметры :",LR.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",LR.best_score_)
```

```python
no_balance_results.append(('Логистическая регрессия: F1', LR.best_score_))
```

<b>Вывод: </b> логистическая регрессия плохо себя показала, имеет крайне низкую F1 - меру.

#### Случайный лес

```python

clf = RandomForestClassifier()
parametrs = { 'n_estimators': range (10, 51, 10),
              'max_depth': range (5,12, 2),
              'min_samples_leaf': range (1,30,10),
              'min_samples_split': range (2,50,10),
            'random_state':[42]}

RFC = GridSearchCV(clf, parametrs, cv=3, scoring = 'f1', verbose = 2)
RFC.fit(X_train, y_train)
```

```python

print("Лучшие гиперпараметры :",RFC.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",RFC.best_score_)
no_balance_results.append(('Случайный лес: F1', RFC.best_score_))
```

<b>Вывод: </b> случайный лес показал себя намного лучше, f1-мера уже 57% на несбалансированной выборке

#### Дерево решений

```python

tree = DecisionTreeClassifier()
parametrs = { 'max_depth': range (1, 10, 1),
              'min_samples_leaf': range (1,30,5),
              'min_samples_split': range (2,50,5),
            'random_state':[42]}

DTC = GridSearchCV(tree, parametrs, cv=3, scoring = 'f1', verbose = 2)
DTC.fit(X_train, y_train)
```

```python

print("Лучшие гиперпараметры :",DTC.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",DTC.best_score_)
no_balance_results.append(('Дерево решений: F1', DTC.best_score_))
```

<b>Вывод: </b> дерево решений показало себя очень хорошо. Результаты сравними со случайным лесом, но обучение модели занимает гораздо меньше времени.

#### К ближайших соседей

```python
model = KNeighborsClassifier()

parametrs = { 'n_neighbors': range (5, 200, 20),
              'weights': ('uniform', 'distance'),
               }

KNC = GridSearchCV(model, parametrs, cv=3, scoring = 'f1', verbose = 2)
KNC.fit(X_train, y_train)
```

```python
print("Лучшие гиперпараметры :",KNC.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",KNC.best_score_)
no_balance_results.append(('К-соседей: F1', KNC.best_score_))
```

<b>Вывод: </b> KNeighborsClassifier показал неплохой результат, но хуже чем предыдущие модели

## Борьба с дисбалансом

```python
#Определяем переменную, которая будет хранить результаты разных моделей для 
#балансированной выборки, чтобы потом сравнить результаты
balance_results = []
```

### Гиперпараметр class_weight
Будем использовать гиперпараметр class_weight для балансировки данных. 
#### Логистическая регрессия

```python
param_grid = {"C": np.arange(0.0, 1.0, 0.1), "penalty":["l2"], 'class_weight':['balanced'], 'random_state':[42]}
LR2 = GridSearchCV( LogisticRegression(), param_grid, 
    cv = 3, scoring = 'f1', verbose = 2)

LR2.fit(X_train, y_train)

print("Лучшие гиперпараметры :",LR2.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",LR2.best_score_)
balance_results.append(('Логистическая регрессия: F1', LR2.best_score_))
```

#### Случайный лес 

```python
clf = RandomForestClassifier()
parametrs = { 'n_estimators': range (10, 51, 10),
              'max_depth': range (5,12, 2),
              'min_samples_leaf': range (1,30,10),
              'min_samples_split': range (2,50,10),
            'class_weight':['balanced'],
            'random_state':[42]}

RFC2 = GridSearchCV(clf, parametrs, cv=3, scoring = 'f1', verbose = 2)
RFC2.fit(X_train, y_train)
```

```python
print("Лучшие гиперпараметры :",RFC2.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",RFC2.best_score_)
balance_results.append(('Случайный лес: F1', RFC2.best_score_))
```

#### Дерево решений

```python
tree = DecisionTreeClassifier()
parametrs = { 'max_depth': range (1, 10, 1),
              'min_samples_leaf': range (1,30,5),
              'min_samples_split': range (2,50,5),
            'class_weight':['balanced'],
            'random_state':[42]}

DTC2 = GridSearchCV(tree, parametrs, cv=3, scoring = 'f1', verbose = 2)
DTC2.fit(X_train, y_train)
```

```python
print("Лучшие гиперпараметры :",DTC2.best_params_)
print("F1_score для лучшей модели на обучающей выборке:",DTC2.best_score_)
balance_results.append(('Дерево решений: F1', DTC2.best_score_))
```

### Upsampling

```python
#Определяем переменную, которая будет хранить результаты разных моделей для 
#балансированной выборки, чтобы потом сравнить результаты
balance_results2 = []
```

Для борьбы с дисбалансом используем метод upsampling, то есть увеличим обучающую выборку таким образом, чтобы количество нулей и единиц стало равным

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=42)    
    return features_upsampled, target_upsampled
```

Так как при upsampling нельзя проводить кросс-валидацию, нужно снова разделить выборку. Так как та часть, что является тестовой - неприкосновенно, разделим обучающую на обучающую меньшего размера и валидационную.

### Downsample
Уменьшение выборки для борьбы с дисбалансом

```python
#Определяем переменную, которая будет хранить результаты разных моделей для 
#балансированной выборки, чтобы потом сравнить результаты
balance_results3 = []
```

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=42)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=42)] + [target_ones])    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=42)    
    return features_downsampled, target_downsampled
```
